```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Ejercicio 1: Carga y Exploración Inicial del Dataset

df = pd.read_csv('clima_talca_enero2024.csv')

display(df.head(5))
display(df.tail(3))
df.info()
display(df.describe())
print(df.shape)
```

P1.1 ¿Cuántas filas y columnas tiene el dataset?
31 filas y 6 columnas

P1.2 ¿Qué columnas contienen datos de tipo float? ¿Cuál es la temperatura máxima promedio según describe()?
Según info(), las columnas de tipo float son: temp_max, temp_min, humedad_pct y precipitacion_mm. Según describe(), la temperatura máxima promedio (mean) es aproximadamente 32.24 °C 

P1.3 ¿Qué información relevante entrega df.info() que NO entrega df.describe()?
df.info() entrega los tipos de datos de cada columna y el conteo exacto de valores no nulos, lo que ayuda a detectar rápidamente datos faltantes, df.describe() solo muestra cálculos matemáticos de las columnas numéricas e ignora los nulos

```python
#Ejercicio 2: Detección de Valores Faltantes 
conteo_nulos = df.isnull().sum()
print(conteo_nulos)

porcentaje_nulos = (df.isnull().sum() / len(df)) * 100
print(porcentaje_nulos)

columnas_criticas = porcentaje_nulos[porcentaje_nulos > 15]
print("Columnas con más del 15% de nulos")
print(columnas_criticas)
```

P2.1 ¿Cuántos valores faltantes tiene el dataset en total?
Sumando todos los nulos, el dataset tiene un total de 17 valores faltantes

P2.2 ¿Qué columnas superan el 15% de nulos? ¿Qué implicancia tiene esto para el análisis?
Las columnas temp_max y precipitacion_mm superan el 15%. Si rellenamos estos datos de forma incorrecta, podríamos sesgar los promedios mensuales e inventar información que no ocurrió en la realidad

P2.3 ¿Por qué es importante calcular el porcentaje y no sólo el conteo absoluto de nulos?
Porque depende del total de datos que estamos tomando en cuenta, a menor datos totales mayor % representa cada valor nulo.

```python
#Ejercicio 3: Limpieza del DataFrame 
df_limpio = df.copy()

df_limpio['temp_max'] = df_limpio['temp_max'].fillna(df_limpio['temp_max'].mean())
df_limpio['temp_min'] = df_limpio['temp_min'].fillna(df_limpio['temp_min'].mean())

df_limpio['humedad_pct'] = df_limpio['humedad_pct'].fillna(df_limpio['humedad_pct'].median())
df_limpio['precipitacion_mm'] = df_limpio['precipitacion_mm'].fillna(0.0)
df_limpio = df_limpio.dropna(subset=['condicion'])

#Verificar que ya no queden nulos

print(df_limpio.isnull().sum())
```

P3.1 ¿Por qué se usó fillna(mean()) para temp_max y fillna(median()) para humedad_pct? ¿Cuándo conviene usar uno u otro?
Se usa el promedio (mean) en la temperatura porque sus valores cambian de forma gradual. Se usa la mediana (median) en la humedad porque evita que los valores atípicos (como un día de lluvia extrema) distorsionen la realidad mensual.

P3.2 ¿Cuántas filas se eliminaron al aplicar dropna(subset=['condicion'])? ¿Es aceptable esta pérdida?
Se eliminaron solo 2 filas. Es una pérdida mínima y totalmente aceptable, ya que la condición climática es un dato categórico (texto) y no se puede inventar ni promediar matemáticamente.

P3.3 ¿Qué sucede si se aplica df.dropna() sin parámetros? ¿Cuántas filas quedarían? Pruébalo.
Si se usa sin parámetros, elimina cualquier fila que tenga al menos un nulo. Quedarían solo 16 filas, lo que sería un error grave porque estaríamos borrando la mitad del mes por culpa de datos faltantes aislados.

```python
#Ejercicio 4: Selección y Filtrado de Datos
display(df_limpio[['fecha', 'temp_max', 'condicion']].head())


display(df_limpio.iloc[:5])
display(df_limpio.loc)
display(df_limpio[df_limpio['temp_max'] >= 34])
display(df_limpio[df_limpio['condicion'] == 'Lluvioso'])
display(df_limpio[(df_limpio['temp_max'] > 30) & (df_limpio['humedad_pct'] > 50)])


df_limpio['amplitud_termica'] = df_limpio['temp_max'] - df_limpio['temp_min']
```

P4.1 ¿Cuántos días tuvo una temperatura máxima igual o superior a 34°C? ¿Qué condición climática predominó esos días?
Hubo 6 días con 34°C o más. La condición climatica que predominó  fue 'Soleado'

P4.2 ¿Cuál es la amplitud térmica promedio del mes? ¿Qué día tuvo la mayor amplitud?
La amplitud térmica promedio es de 14.7 °C. El día con la mayor amplitud térmicafue el 2024-01-06 con 16.5°C

P4.3 ¿Por qué se utilizan paréntesis adicionales en el filtro compuesto? ¿Qué ocurre si se omiten?
Se usan paréntesis porque en Python los operadores lógicos & y | tienen mayor prioridad matemática que los operadores de comparación (>). Si se omiten, Pandas intenta ejecutar primero la operación 30 & df_limpio['humedad_pct'], lo cual arroja un error inmediato de sintaxis.

```python
#Ejercicio 5: Análisis Integrador y Estadísticas Descriptivas

dia_mas_caluroso = df_limpio.loc[df_limpio['temp_max'].idxmax()]
print(f"El día más caluroso fue: {dia_mas_caluroso['fecha']} con {dia_mas_caluroso['temp_max']}°C")

dia_lluvia = df_limpio.loc[df_limpio['precipitacion_mm'].idxmax()]
print(f"Día con mayor precipitación: {dia_lluvia['fecha']} con {dia_lluvia['precipitacion_mm']} mm")


print("Dias por condición climática:")
print(df_limpio['condicion'].value_counts())

print(f"\n Temp Máxima promedio: {df_limpio['temp_max'].mean():.2f}°C")
print(f"   Temp Mínima promedio: {df_limpio['temp_min'].mean():.2f}°C")

print(f"\n Cantidad de filas duplicadas: {df_limpio.duplicated().sum()}")

#Resumen final
display(df_limpio.describe())
```

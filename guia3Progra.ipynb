```python
from functools import total_ordering
@total_ordering

class Producto:
    def __init__(self, nombre : str, precio : float, stock : int):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    def __str__(self):
        return f"{self.nombre} - ${self.precio} ({self.stock} unidades)"

    def __repr__(self):
        return f"Producto('{self.nombre}',{self.precio},{self.stock})"

    def __eq__(self, otro):
        return self.precio == otro.precio

    def __lt__(self, otro):
        return self.precio < otro.precio

    def __add__(self, otro):
        return self.stock + otro.stock

    def __mul__(self,n):
        return Producto(self.nombre, self.precio * n, self.stock)

p1 = Producto('Arroz', 1200, 50)
p2 = Producto('Arroz', 900, 30)
p3 = Producto('Leche', 1200, 20)


print(p1) # Arroz – $1200 (50 unidades)
print(repr(p2)) # Producto('Arroz', 900, 30)
print(p1 == p3) # True → mismo precio
print(p2 < p1) # True → 900 < 1200
print(sorted([p1, p2, p3])) #ordenado por precio

combinado = p1 + p2 # stock = 80
print(combinado) # Arroz – $1200 (80 unidades)

oferta = p1 * 0.8 # precio con 20% descuento
print(oferta) # Arroz – $960.0 (50 unidades)
```

```python
class CuentaBancaria:
    def __init__(self, titular: str, saldo: float):
        self.titular = titular
        self.saldo = saldo 
        
    @property
    def titular(self) -> str:
        return self._titular
        
    @titular.setter
    def titular(self, nombre: str):
        if len(nombre) < 3:
            raise ValueError("El nombre de titular debe tener al menos 3 caracteres")
        self._titular = nombre

    @property
    def saldo(self) -> float:
        return self._saldo
        
    @saldo.setter
    def saldo(self, valor: float):
        if valor < 0:
            raise ValueError(f"El saldo no puede ser negativo")
        self._saldo = valor

    def depositar(self, monto: float):
        if monto > 0:
            self.saldo = self.saldo + monto

    def retirar(self, monto: float):
        if self.saldo - monto < 0:
            raise ValueError("No existe saldo suficiente")
        self.saldo = self.saldo - monto 
    @property
    def aprobada(self) -> bool:
        return self.saldo >= 10000

    def __str__(self):
        return f"Cuenta de {self.titular} | Saldo: ${self.saldo}"

    def __repr__(self):
        return f"CuentaBancaria(titular='{self.titular}', saldo={self.saldo})"



cuenta = CuentaBancaria('Ana López', 50000)
print(cuenta) # Cuenta de Ana López | Saldo: $50000

cuenta.depositar(20000)
print(cuenta.saldo) # 70000

cuenta.retirar(15000)
print(cuenta) # Cuenta de Ana López | Saldo: $55000

# Validaciones
try:
    cuenta.saldo = -1000 # debe lanzar ValueError
except ValueError as e:
    print(e) # Saldo no puede ser negativo
    
try:
    CuentaBancaria('AB', 0) # titular muy corto
except ValueError as e:
    print(e)
    
print(cuenta.aprobada) # True
```

```python
import math

class Figura:
    def __init__(self, color: str):
        self.color = color

    def area(self) -> float:
        raise NotImplementedError
        
    def perimetro(self) -> float:
        raise NotImplementedError

    def __str__(self):
        return f"Figura {self.color} | Area: {self.area():.2f} | Perimetro: {self.perimetro():.2f}"
        
    def __repr__(self):
        return f"Figura(color='{self.color}')"

    def __lt__(self, otro):
        return self.area() < otro.area()

class Circulo(Figura):
    def __init__(self, color: str, radio: float):
        super().__init__(color) 
        if radio <= 0:
            raise ValueError("El radio debe ser mayor a 0")
        self.radio = radio

    def area(self) -> float:
        return math.pi * self.radio ** 2
        
    def perimetro(self) -> float:
        return 2 * math.pi * self.radio
        
    def __repr__(self):
        return f"Circulo('{self.color}', {self.radio})"

class Rectangulo(Figura):
    def __init__(self, color: str, ancho: float, alto: float):
        super().__init__(color) 
        if ancho <= 0 or alto <= 0:
            raise ValueError("Ancho y alto deben ser mayores a 0")
        self.ancho = ancho
        self.alto = alto

    def area(self) -> float:
        return self.ancho * self.alto
        
    def perimetro(self) -> float:
        return 2 * (self.ancho + self.alto)

    @property
    def es_cuadrado(self) -> bool:
        return self.ancho == self.alto
        
    def __repr__(self):
        return f"Rectangulo('{self.color}', {self.ancho}, {self.alto})"


c = Circulo('rojo', 5)
r = Rectangulo('azul', 4, 6)
cuad = Rectangulo('verde', 5, 5)
    
print(c) # Figura roja | Área: 78.54 | Perímetro: 31.42
print(r) # Figura azul | Área: 24 | Perímetro: 20

print(cuad.es_cuadrado) # True
print(r.es_cuadrado) # False

# Polimorfismo
figuras = [c, r, cuad]
for f in figuras:
    print(f'Área: {f.area():.2f}')
# Ordenar por área (requiere __lt__)
# Bonus: agrega __lt__ basado en área y ordénalas
```

```python
class Inventario:
    def __init__(self):
        self._productos = []

    def agregar(self, producto):
        self._productos.append(producto)

    def __len__(self) -> int:
        return len(self._productos)

    def __getitem__(self, i: int):
        return self._productos[i]

    def __contains__(self, producto):
        return producto in self._productos

    def __iter__(self):
        return iter(self._productos)

    def __str__(self):
        return "\n".join(str(p) for p in self._productos)

    def __repr__(self):
        return f"Inventario(n_productos={len(self._productos)})"

    @property
    def total_valor(self):
        return sum(p.precio * p.stock for p in self._productos)

    @property
    def mas_caro(self):
        if not self._productos:
            return None
        return max(self._productos, key=lambda p: p.precio)

    def buscar(self, nombre: str) -> list:
        return [p for p in self._productos if nombre.lower() in p.nombre.lower()]



inv = Inventario()
inv.agregar(Producto('Arroz', 1200, 50))
inv.agregar(Producto('Leche', 950, 30))
inv.agregar(Producto('Pan', 1800, 100))
inv.agregar(Producto('Panetón', 3500, 10))

print(len(inv)) # 4
print(inv[0]) # Arroz – $1200 (50 unidades)

p_buscar = Producto('Leche', 950, 30)
print(p_buscar in inv) # True

for p in inv:
    print(p)
    
print(inv.total_valor) # 1200*50 + 950*30 + 1800*100 + 3500*10 = ...
print(inv.mas_caro) # Panetón – $3500 (10 unidades)

resultados = inv.buscar('pan')
for r in resultados:
    print(r) # Pan y Panetón
```

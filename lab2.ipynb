```python
#Actividad 1 – Uso de lambda
doble = lambda x: (x **2 + 2*x + 1)
#Probar la función con valores 2, 4 y 6.
print(doble(2))
print(doble(4))
print(doble(6))

#los valores obtenidos son correctos segun la formula
```

```python
#Actividad 2 – Transformación con map()
velocidades = [10,20,30,40,50]
#Convertir las velocidades de km/h a m/s:
metroseg = list(map(lambda x: x/3.6, velocidades))
metroseg

#las velocidades pasan a m/s
```

```python
#Actividad 3 – Filtrado con filter()
valores = [2,5,8,11,14,17,20]
#Filtrar valores pares en la lista:
pares = list(filter(lambda x: x % 2 ==0, valores))
pares

#se extraen los valores pares de la lista
```

```python
#Actividad 4 – Agregación con reduce()
from functools import reduce
numeros = [2,3,4,5]
#Calcular el producto total de los valores
producto = reduce(lambda x,y: x*y, numeros)
producto

#se calcula el producto de los valores de la lista 
```

```python
#Actividad 5 – Pipeline de Procesamiento
from functools import reduce
data = [12,15,22,35,41,8,3,55,62,90]
#Eliminar valores menores a 10
#Multiplicar cada valor por 2
#Calcular la suma total
resultado = reduce(lambda x,y: x+y,
map(lambda x: x*2,
filter(lambda x: x>=10, data)))
resultado

#el resultado es lo esperado 
```

```python
#Actividad 6 – Simulación de Big Data
from functools import reduce
import random
data = [random.randint(0,100) for _ in range(1000000)]
#Filtrar valores mayores a 50
#Elevar los valores al cuadrado
#Calcular el promedio final
datos_procesados = list(map(lambda x: x**2, filter(lambda x: x>50, data)))
resultado = reduce(lambda x, y: x+y, datos_procesados) / len(datos_procesados)
resultado

#se aplican correctamente las funciones 
```

```python
#Desafío Avanzado - Simular sensores industriales
from functools import reduce
import random
sensores = [random.uniform(10,80) for _ in range(500000)]
#Eliminar valores fuera del rango 10–70 °C
#Convertir a Kelvin
#Calcular temperatura promedio
sensores_procesados = list(map(lambda x: x + 273.15, filter(lambda x: 10<=x<=70, sensores)))
resultado = reduce(lambda x, y: x + y, sensores_procesados) / len(sensores_procesados)
resultado

#se aplican correctamente las funciones
```

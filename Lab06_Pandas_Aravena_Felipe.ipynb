¿Por qué en Pandas se usan & y | en lugar de and y or para combinar condiciones?

Porque and y or evalúan el valor de verdad de todo el objeto a la vez, lo cual genera un error en estructuras grandes. Los símbolos & y | permiten a Pandas realizar operaciones lógicas "elemento por elemento" (fila por fila) de forma vectorizada.

¿Que ventaja tiene .isin() frente a escribir multiples condiciones con |?

La sintaxis es mucho más limpia, fácil de leer y permite agregar múltiples elementos a una lista sin tener que repetir el nombre de la columna y el operador == varias veces.

2.5

Crea una función Python que reciba como parámetros: region, categoria y monto_minimo,
y retorne el DataFrame filtrado según esos criterios. Se usan 3 diferentes parametros y se usa head para mostrar las 3 primeras filas

```python
def filtrar_ventas(region, categoria, monto_minimo):
    return df[
        (df['region'] == region) & 
        (df['categoria'] == categoria) & 
        (df['monto_bruto'] >= monto_minimo)
    ]

display(filtrar_ventas('Norte', 'Laptop', 1000).head(3))
display(filtrar_ventas('Sur', 'Tablet', 500).head(3))
display(filtrar_ventas('Centro', 'Laptop', 800).head(3))
```

3.1

Agrupa el dataset por 'region' y calcula para la columna monto_bruto: suma total,
promedio, máximo, mínimo y conteo de transacciones. Usa .agg() con nombres de
columnas personalizados y aplica. round(2).

```python
resumen_region = df.groupby('region')['monto_bruto'].agg(
    Suma_Total='sum',
    Promedio='mean',
    Maximo='max',
    Minimo='min',
    Conteo='count'
)
display(resumen_region)
```


3.2

Sobre el resultado anterior, usa .reset_index() y ordena de mayor a menor según el total de

```python
resumen_region.round(2).reset_index().sort_values(by='Suma_Total', ascending=False)
```

3.5

Agrupa por categoría y calcula el precio_unit promedio y la cantidad total. Ordena por
precio promedio descendente e interpreta el resultado: ¿que categoría tiene el mayor
ticket promedio?

```python
resumen_cat = df.groupby('categoria').agg(
    precio_promedio=('precio_unit', 'mean'),
    cantidad_total=('cantidad', 'sum')
).sort_values(by='precio_promedio', ascending=False)
display(resumen_cat)
```

¿Para qué sirve .reset_index() después de un groupby? ¿Qué sucede si no lo aplicas? 

Sirve para aplanar el índice jerárquico que crea el groupby y convertirlo de nuevo en columnas normales del DataFrame. Si no se aplica, las variables agrupadas (como la región o el mes) quedan "atrapadas" como el índice de la tabla, dificultando su uso en filtros posteriores.

¿Cuándo es preferible .agg() con nombres vs. usar .sum(), .mean() directamente?

Es preferible usar .agg() cuando necesitas calcular múltiples métricas estadísticas distintas (como suma, promedio y máximo) sobre la misma columna en un solo paso, y además te permite asignarles nombres limpios y profesionales al instante.

4.1

Crea una tabla pivote usando pd.pivot_table() con: indice = 'region', columnas = 'mes',
valores = 'monto_bruto', funcion de agregación = suma. Aplica .round(2) al resultado.

```python
pivot_mensual = pd.pivot_table(
    df, 
    index='region', 
    columns='mes', 
    values='monto_bruto', 
    aggfunc='sum'
).round(2)
```

4.2

la tabla pivote, usa .fillna(0) para reemplazar los valores NaN por 0. Explica en
Markdown por que pueden existir NaN en una pivot_table.

```python
pivot_mensual.fillna(0, inplace=True)
```

4.3

Agrega una columna de totales al final de la tabla pivote usando pivot['Total'] =
pivot.sum(axis=1). ¿Cuál es la región con mayor total semestral?

```python
pivot_mensual['Total'] = pivot_mensual.sum(axis=1)
display(pivot_mensual)
```

Aplica formato de estilo con pivot.style.format('${:,.2f}').highlight_max(axis=0, color='#E8F8F5')

```python
display(pivot_mensual.style.format('${:.2f}').highlight_max(axis=0, color='#E8F8F5'))
```

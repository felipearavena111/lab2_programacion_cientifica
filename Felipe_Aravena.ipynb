```python
class Empleado:
    def __init__(self, nombre, empresa, salario_base):
        self.nombre = nombre
        self.empresa = empresa
        self.salario_base = salario_base

    def calcular_salario(self):
        return self.salario_base

    def descripcion(self):
        return f"Empleado: {self.nombre} - Salario: ${self.calcular_salario()}"


class Gerente_Ventas(Empleado):
    def __init__(self, nombre, empresa, salario_base, bono_ventas):
        super().__init__(nombre, empresa, salario_base)
        self.bono_ventas = bono_ventas

    def calcular_salario(self):
        return self.salario_base + self.bono_ventas
        
    def descripcion(self):
        return f"Gerente Ventas: {self.nombre} - Salario: ${self.calcular_salario()}"


class Obrero_PorHora(Empleado):
    def __init__(self, nombre, empresa, horas_trabajadas, valor_hora):
        super().__init__(nombre, empresa, 0)
        self.horas_trabajadas = horas_trabajadas
        self.valor_hora = valor_hora

    def calcular_salario(self):
        return self.horas_trabajadas * self.valor_hora
        
    def descripcion(self):
        return f"Obrero PorHora: {self.nombre} - Salario: ${self.calcular_salario()}"
```

```python
gerente = Gerente_Ventas("Ana Pérez", "UA", 1500000, 800000)
obrero = Obrero_PorHora("Pedro Soto", "UCM", 160, 4000)

print(gerente.descripcion())
print(obrero.descripcion())
```

```python
from abc import ABC, abstractmethod

class Figura(ABC):    
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimetro(self):
        pass
        
    def describir(self):
        return f"{self.__class__.__name__} - Area: {self.area():.2f}, Perimetro: {self.perimetro():.2f}"

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio
        
    def area(self):
        return 3.146 * (self.radio ** 2)
        
    def perimetro(self):
        return 2 * 3.1416 * self.radio

class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
        
    def area(self):
        return self.base * self.altura
        
    def perimetro(self):
        return 2 * (self.base + self.altura)

class Triangulo(Figura):
    def __init__(self, base, altura, hip):
        self.base = base
        self.altura = altura
        self.hip = hip
        
    def area(self):
        return (self.base * self.altura) / 2
        
    def perimetro(self):
        return self.base + self.altura + self.hip

def calcular_total(figuras : list):
    return sum(f.area() for f in figuras)

```

```python
from abc import ABC, abstractmethod

class Metodo_Pago(ABC):
    @abstractmethod
    def procesar(self, monto):
        pass

    @abstractmethod
    def reembolsar(self, monto):
        pass

class PagoTarjeta(Metodo_Pago):
    def __init__(self, numero_tarjeta, titular):
        self.numero_tarjeta = numero_tarjeta[-4:] 
        self.titular = titular
        
    def procesar(self, monto):
        return f"Procesando ${monto} con tarjeta terminada en {self.numero_tarjeta} (Titular: {self.titular})"
        
    def reembolsar(self, monto):
        return f"Reembolsando ${monto} a la tarjeta terminada en {self.numero_tarjeta}"

class Pago_Transferencia(Metodo_Pago):
    def __init__(self, cuenta_origen, cuenta_destino):
        self.cuenta_origen = cuenta_origen
        self.cuenta_destino = cuenta_destino
        
    def procesar(self, monto):
        return f"Transfiriendo ${monto} desde cuenta {self.cuenta_origen} a {self.cuenta_destino}"
        
    def reembolsar(self, monto):
        return f"Reversando transferencia de ${monto} hacia la cuenta {self.cuenta_origen}"

class PagoEfectivo(Metodo_Pago):
    def procesar(self, monto):
        return f"Recibiendo pago en efectivo por ${monto} en caja"
        
    def reembolsar(self, monto):
        return f"Entregando reembolso en efectivo por ${monto} en caja"

def ejecutar_pago(metodo: Metodo_Pago, monto: float):
    print(metodo.procesar(monto))

```

```python
pagos = [
    PagoTarjeta("1234-5678-9012-3456", "Ana Pérez"),
    Pago_Transferencia("21432542", "21532452"),
    PagoEfectivo()
]

for i in pagos:
    ejecutar_pago(i, 50000)
```
